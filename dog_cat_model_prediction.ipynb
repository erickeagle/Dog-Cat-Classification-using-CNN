```python
import cv2 
import tensorflow as tf
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from keras.models import load_model
```

```python
CATEGORIES=['Dog','Cat']
```

```python
def prepare(image):
    img_size=100
    img_array=cv2.imread(image,cv2.IMREAD_GRAYSCALE)
    new_array=cv2.resize(img_array,(img_size,img_size))
    return new_array.reshape(-1,img_size,img_size,1)
```

```python
model = tf.keras.models.load_model(r"K:\kagglecatsanddogs_3367a\Dogs_vs_Cats_100.model")

```

```python
image=r'K:\kagglecatsanddogs_3367a\TestImages\119.jpg'
```

```python
prediction=model.predict([prepare(image)/255.0])
print(prediction)
print(round(prediction[0][0]))
print(CATEGORIES[int(round(prediction[0][0]))])
```

```python
img=mpimg.imread(image)
imgplot=plt.imshow(img)
plt.title(CATEGORIES[int(prediction[0][0])])
plt.show()
```
